```python
def flatten_files(extract_dir, final_dir):
  os.makedirs(extract_dir, exist_ok=True)
  os.makedirs(final_dir, exist_ok=True)
  for root, dirs, files in tqdm(os.walk(extract_dir)):
      for file in files:
          src_file = os.path.join(root, file)
          dest_file = os.path.join(final_dir, file)

          # If a file with the same name exists, rename it to avoid conflicts
          if os.path.exists(dest_file):
              base, ext = os.path.splitext(file)
              dest_file = os.path.join(final_dir, f"{base}_copy{ext}")
              print(f"Warning: Renaming file {base}{ext} -> {base}_copy{ext}")

          shutil.move(src_file, dest_file)
```
